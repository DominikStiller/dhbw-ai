```python
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd

sb.set(style="white", palette="muted", color_codes=True)
```

```python
data = pd.read_csv('dataset_full.csv')
tarif_order = ['Tarif A', 'Tarif B', 'Tarif C', 'Tarif D', 'ablehnen']
```

```python
data
```

```python
sb.catplot(x='Versicherungstarif', data=data, kind='count')
```

### Alter

```python
g = sb.catplot(x='Alter', data=data, kind='count')
plt.title('Altersverteilung')
g.set_xticklabels([str(i) if i % 10 == 0 else '' for i in range(data['Alter'].min(), data['Alter'].max())])
```

```python
sb.catplot(x='Alter', y='Versicherungstarif', order=tarif_order, data=data, kind='swarm')
```

```python
sb.catplot(x='Alter', y='Familienstand', data=data, kind='swarm')
```

```python
sb.catplot(x='Alter', y='Beruf', data=data, kind='swarm')
```

```python
sb.relplot(x='Alter', y='Jahresgehalt', data=data)
```

* Alter -> Jahresgehalt
* Alter in Buckets alle 20 Jahre

```python
sb.countplot(x='Familienstand', hue='Versicherungstarif', hue_order=tarif_order, data=data)
```

```python
sb.countplot(x='Kinder', hue='Versicherungstarif', hue_order=tarif_order, data=data)
```

```python
sb.countplot(x='Beruf', hue='Versicherungstarif', hue_order=tarif_order, data=data)
```

```python
sb.catplot(x='Jahresgehalt', y='Versicherungstarif', order=tarif_order, data=data, kind='swarm')
```

```python
sb.catplot(x='Jahresgehalt', y='Beruf', data=data, kind='swarm')
```

```python
sb.countplot(x='Geschlecht', hue='Versicherungstarif', hue_order=tarif_order, data=data)
```

```python
sb.catplot(x='Jahresgehalt', y='Geschlecht', data=data, kind='swarm')
```

```python
data.groupby('Geschlecht')['Jahresgehalt'].mean()
```
